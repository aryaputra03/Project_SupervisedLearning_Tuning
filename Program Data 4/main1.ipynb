```python
from sklearn.preprocessing import LabelEncoder


labencod = LabelEncoder()
df['label'] = labencod.fit_transform(df['label'])
```

```python
df_new = df.copy()

for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.3 * IQR
    lower_bound = Q1 - 1.3 * IQR
    
    print(col)
    print('Upper Limit :', upper_bound)
    print('Lower Limit :', lower_bound)
    print("---")
    
    # Keep only rows within the bounds for this column
    df_new = df_new[(df_new[col] >= lower_bound) & (df_new[col] <= upper_bound)]
```

```python


fig,axs = matplotlib.pyplot.subplots(nrows=6,ncols=5,figsize=(20,12))

for i,j in enumerate(df.columns):
    row,col = divmod(i,5)
    seaborn.boxplot(df_new[j],ax=axs[row,col])
    axs[row,col].set_title(j)
matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```

```python
x_n = df.drop(columns='label')
y_n = df['label']
 
x_o = df_new.drop(columns='label')
y_o = df_new['label']

print(f'{x_n.shape}/{y_n.shape} : {x_o.shape}/{y_o.shape}')
```

```python
from sklearn.model_selection import train_test_split


xtrain_n,xtest_n,ytrain_n,ytest_n = train_test_split(x_n,y_n,random_state=42,test_size=0.2)

xtrain_o,xtest_o,ytrain_o,ytest_o = train_test_split(x_o,y_o,random_state=42,test_size=0.2)
```

# CB

```python
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

model_cb_n = CatBoostClassifier(random_state=42,verbose=0)
model_cb_n.fit(xtrain_n,ytrain_n)
pred_cb_n = model_cb_n.predict(xtest_n)

model_cb_o = CatBoostClassifier(random_state=42,verbose=0)
model_cb_o.fit(xtrain_o,ytrain_o)
pred_cb_o = model_cb_o.predict(xtest_o)

print(f" DATA N -> ACC = {accuracy_score(ytest_n,pred_cb_n)} / F1 = {f1_score(ytest_n,pred_cb_n,average='macro')}")
print(f" DATA O -> ACC = {accuracy_score(ytest_o,pred_cb_o)} / F1 = {f1_score(ytest_o,pred_cb_o,average='macro')}")  
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier


model_rf_n = RandomForestClassifier(random_state=42) 
model_rf_n.fit(xtrain_n,ytrain_n)
pred_rf_n = model_rf_n.predict(xtest_n)

model_rf_o = RandomForestClassifier(random_state=42)
model_rf_o.fit(xtrain_o,ytrain_o)
pred_rf_o = model_rf_o.predict(xtest_o)


print(f"D_Normal ACC = {accuracy_score(ytest_n,pred_rf_n)} / F1 = {f1_score(ytest_n,pred_rf_n,average='macro')}")
print(f"D_OUT ACC = {accuracy_score(ytest_o,pred_rf_o)} / F1 = {f1_score(ytest_o,pred_rf_o,average='macro')}")
```
